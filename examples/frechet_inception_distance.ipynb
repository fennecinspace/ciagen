```python
import torch
from torchvision.datasets import CIFAR10
import numpy as np

from tqdm import tqdm
```

```python
real_dummy = torch.rand((1000, 100))
synthetic_dummy = torch.rand((1000, 100))

fid = FID()
fid_score = fid.instant_score(real_dummy, synthetic_dummy, already_transformed=True)
print(fid_score)
```

```python
real_data = CIFAR10(root="./data", train=True, download=True)
synthetic_data = CIFAR10(root="./data", train=False, download=True)

LIMIT = len(real_data) if len(real_data) < len(synthetic_data) else len(synthetic_data)
N = 100 if LIMIT > 1000 else LIMIT / 2
L = 1000

index = np.random.choice(LIMIT, N, replace=False)
real_dataset = [real_data[i][0] for i in index]
synthetic_dataset = [synthetic_data[i][0] for i in index]
```

```python
fid_frechet = FID()
fid_frechet_score = fid_frechet.instant_score(real_dataset, synthetic_dataset)
fid_frechet_score
```

```python
fid_wasserstein = FID(use_wasserstein=True)
fid_wasserstein_score = fid_wasserstein.instant_score(real_dataset, synthetic_dataset)
fid_wasserstein_score
```
