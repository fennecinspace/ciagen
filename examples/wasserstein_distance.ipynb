```python
mean1 = np.random.rand(1000,)
var1 = np.random.rand(1000,1000)
var1 = np.matmul(var1, var1.T)

mean2 = np.random.rand(1000,)
var2 = np.random.rand(1000,1000)
var2 = np.matmul(var2, var2.T)

usample = np.random.multivariate_normal(mean1, var1, 1000)
vsample = np.random.multivariate_normal(mean2, var2, 1000)
```

```python
a = np.linalg.norm(mean1 - mean2, ord=2)
b = a ** 2
print(a, b)
```

```python
res = wasserstein_distance_gaussian_version(mean1, var1, mean2, var2)
res2 = wasserstein_distance_multi_dimensional(usample, vsample, as_expectance=True)

print(f"Gaussian version: {res} \nMulti-dimensional version: {res2}")
```
